```python
vid=cv2.VideoCapture(0)
while True:
    flag,img=vid.read()
    if flag:
        
        cv2.imshow('home',img)
        key=cv2.waitKey(1)
        if key==ord('q'): ##for stop the camera ligth which is acquire by this program
           break
    else:
        break

vid.release()
cv2.destroyAllWindows()
```
